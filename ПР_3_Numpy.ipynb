```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import rayleigh
```

# Распределение Рэлея

```python
x = np.linspace(scipy.stats.rayleigh.ppf(0.01),
                scipy.stats.rayleigh.ppf(0.99), 100)

fig, ax = plt.subplots(1, 1)

rv = rayleigh()
r = rayleigh.rvs(size=1000)

ax.plot(x, rv.pdf(x), label='pdf')
ax.hist(r, density=True, histtype='stepfilled', alpha=0.3)
ax.legend(loc='best', frameon=False)
plt.show()
```
