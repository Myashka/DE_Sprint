# Палиндром строки

```python
def palindrop(x: str) -> bool:
    x = x.replace(' ', '')
    return x == x[::-1]
```

```python
x = input('x = ')
print(palindrop(x))
```

# Перевод арабского числа в римское

```python
def intToRoman(num: int) -> str:
    m = ["", "M", "MM"]
    c = ["", "C", "CC", "CCC", "CD", "D",
         "DC", "DCC", "DCCC", "CM "]
    x = ["", "X", "XX", "XXX", "XL", "L",
         "LX", "LXX", "LXXX", "XC"]
    i = ["", "I", "II", "III", "IV", "V",
         "VI", "VII", "VIII", "IX"]
  
    thousands = m[num // 1000]
    hundreds = c[(num % 1000) // 100]
    tens = x[(num % 100) // 10]
    ones = i[num % 10]
  
    ans = (thousands + hundreds +
           tens + ones)
  
    return ans
```

```python
x = int(input('x = '))
print(intToRoman(x))
```

# Валидность скобок

```python
def check(string):
    brackets_open = ('(', '[', '{', '<')
    brackets_closed = (')', ']', '}', '>')
    stack = []
    for i in string:
        if i in brackets_open:
            stack.append(i)
        if i in brackets_closed:    
            if len(stack) == 0:
                return False
            index = brackets_closed.index(i)
            open_bracket = brackets_open[index]
            if stack[-1] == open_bracket:
                stack = stack[:-1]  
            else:
                return False  
    return (not stack)
```

```python
x = input('x = ')
print(check(x))
```

# Умножить два бинарных числа в формате строк

```python
x1 = int(input('x1 = '), 2)
x2 = int(input('x2 = '), 2)
```

```python
def to_bin(a: int) -> str:

    i = 1
    steps = 0
    output = ''
    
    # находим наибольшую степень двойки
    while i <= a:
        steps += 1
        i *= 2
        
    
    while steps != 0:
        if i < a:
            output += '1'
            a -= i
            i /= 2
            steps -= 1
        elif i == a:
            output += '1'
            output += '0' * (steps)
            break
        else:
            output += '0'
            steps -= 1
            i /= 2
    
    while output[0] == '0':
        output = output[1:]
    
    return output
```

```python
to_bin(x1*x2)
```

```python
print(int(to_bin(x1*x2), 2) == x1*x2)
```
